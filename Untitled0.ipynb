```python
import requests, zipfile, io
import pandas as pd

# 本日分のレース情報ZIPのURL
url = "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=daily"
headers = {"User-Agent": "Mozilla/5.0"}

# ZIPをダウンロード
r = requests.get(url, headers=headers)

# ZIPをメモリ上で開いて、中に入っているファイル名を一覧表示
z = zipfile.ZipFile(io.BytesIO(r.content))
print("ZIPの中のファイル一覧：")
for name in z.namelist():
    print("　", name)
```

```python
# ZIPの中から出馬表CSVを探して読み込む
horse_file = [n for n in z.namelist() if "horselist" in n][0]

# CSVを読み込む（日本語なので文字コードに注意。まずは一般的なものを試す）
with z.open(horse_file) as f:
    horses = pd.read_csv(f, encoding="utf-8")

# 大きさと最初の数行を表示
print("出馬表の大きさ：", horses.shape)
horses.head()
```

```python
# 列名の一覧を表示
print("列の数：", len(horses.columns))
for c in horses.columns:
    print("　", c)
```

```python
import requests, zipfile, io
import pandas as pd

headers = {"User-Agent": "Mozilla/5.0"}

def try_download(url):
    try:
        r = requests.get(url, headers=headers, timeout=30)
        z = zipfile.ZipFile(io.BytesIO(r.content))
        return z.namelist()
    except Exception as e:
        return f"ダメだった: {type(e).__name__}"

# 2026年5月分を、URLの書き方を変えて3パターン試す
candidates = [
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=monthly&k_year=2026&k_month=5",
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=monthly&k_yearMonth=202605",
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/MonthlyRaceDataDownload?k_year=2026&k_month=5",
]

for url in candidates:
    print(url)
    print("　→", try_download(url))
    print()
```

```python
import requests

headers = {"User-Agent": "Mozilla/5.0"}

# いろんなパラメータの組み合わせを試して、応答の「正体」を確認する
candidates = [
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=monthly&k_year=2026&k_month=5",
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=month&k_year=2026&k_month=5",
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=monthly&k_raceDate=2026/05",
    "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=daily&k_raceDate=2026/05/01",
]

for url in candidates:
    r = requests.get(url, headers=headers, timeout=30)
    ctype = r.headers.get("Content-Type", "")
    print("URL:", url)
    print("  状態:", r.status_code, "／ 種類:", ctype, "／ サイズ:", len(r.content), "バイト")
    # ZIPなら先頭が PK で始まる。中身の先頭を少しだけ覗く
    print("  中身の先頭:", r.content[:80])
    print()
```

```python
import requests, zipfile, io
import pandas as pd

headers = {"User-Agent": "Mozilla/5.0"}

def get_today_horses():
    """当日のレース情報ZIPを取得して、出馬表をDataFrameで返す。
       開催がない・データがない場合は None を返す（エラーで止まらない）。"""
    url = "https://www.keiba.go.jp/KeibaWeb/DataDownload/RaceDataDownload?type=daily"
    try:
        r = requests.get(url, headers=headers, timeout=30)
        z = zipfile.ZipFile(io.BytesIO(r.content))
    except zipfile.BadZipFile:
        print("ZIPが取得できませんでした（本日開催なし、またはデータ未生成の可能性）")
        return None

    # 出馬表CSVを探す
    names = [n for n in z.namelist() if "horselist" in n]
    if not names:
        print("出馬表ファイルが見つかりませんでした")
        return None

    with z.open(names[0]) as f:
        horses = pd.read_csv(f, encoding="shift_jis")
    print(f"取得成功：{horses.shape[0]}頭分のデータ")
    return horses

# 実行してみる
horses = get_today_horses()
```
